```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Data

```python
data = 'Ex_Files_Python_Data_Visualization/Exercise_Files/3 Pandas/data/table_i702t60.csv'
```

```python
df =pd.read_csv(data)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
month_num = df.loc[:,'month'].values
interest_paid = df['interest_paid'].to_numpy()
principal_paid = df['principal_paid'].to_numpy()
```

```python
type(interest_paid)
type(principal_paid)
```

```python
plt.plot(month_num,interest_paid,label = 'Interest')
plt.plot(month_num,principal_paid, label= 'Principal')

plt.legend(loc = "right")


plt.xlim(left = 1, right = 70)
plt.ylim(bottom = 0,top=700)

plt.xlabel('month')
plt.ylabel("Dollars")

plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.title("Interest and Principal Paid Each month",fontsize = 15)

```
